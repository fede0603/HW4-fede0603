```python
def minhash(set,n_perm):
    sign = np.zeros([n_perm+1,len(set[0])])
    for i in range(3):
        j = 0
        while set[j][i] == 0:
             j += 1
        sign[0][i] = j+1  
    for b in range(0,n_perm):
        set = np.random.permutation(set)
        print(set)
        for i in range(3):
            j = 0
            while set[j][i] == 0:
                j += 1
            sign[b+1][i] = j+1
    return sign        
```
